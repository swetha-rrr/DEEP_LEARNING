```python
import numpy as np

class Perceptron:
    def __init__(self, input_size):
        self.weights = np.random.rand(input_size + 1) - 0.5  # Random initialization between -0.5 and 0.5
        self.learning_rate = 0.1

    def predict(self, inputs):
        summation = np.dot(inputs, self.weights[1:]) + self.weights[0]
        return 1 if summation > 0 else 0

    def train(self, training_inputs, labels, max_iterations=1000):
        for iteration in range(max_iterations):
            errors = []
            for inputs, label in zip(training_inputs, labels):
                prediction = self.predict(inputs)
                error = label - prediction
                errors.append(error)
                if error != 0:
                    self.weights[1:] += self.learning_rate * error * inputs
                    self.weights[0] += self.learning_rate * error
            num_errors = sum(1 for error in errors if error != 0)
            print("Iteration:", iteration + 1, "Errors:", num_errors)
            if num_errors == 0:
                print("Converged at iteration", iteration + 1)
                break
        else:
            print("Perceptron did not converge within the maximum number of iterations.")

def boolean_and(inputs):
    return int(all(inputs))

def boolean_or(inputs):
    return int(any(inputs))

def boolean_nand(inputs):
    return int(not all(inputs))

def boolean_xor(inputs):
    return int(sum(inputs) % 2 != 0)

training_inputs = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
labels_and = np.array([boolean_and(inputs) for inputs in training_inputs])
labels_or = np.array([boolean_or(inputs) for inputs in training_inputs])
labels_nand = np.array([boolean_nand(inputs) for inputs in training_inputs])
labels_xor = np.array([boolean_xor(inputs) for inputs in training_inputs])

# Perceptron for AND
print("Training Perceptron for AND")
and_perceptron = Perceptron(input_size=2)
and_perceptron.train(training_inputs, labels_and)

# Perceptron for OR
print("\nTraining Perceptron for OR")
or_perceptron = Perceptron(input_size=2)
or_perceptron.train(training_inputs, labels_or)

# Perceptron for NAND
print("\nTraining Perceptron for NAND")
nand_perceptron = Perceptron(input_size=2)
nand_perceptron.train(training_inputs, labels_nand)

# Perceptron for XOR
print("\nTraining Perceptron for XOR")
xor_perceptron = Perceptron(input_size=2)
xor_perceptron.train(training_inputs, labels_xor)
```

```python
import numpy as np
import matplotlib.pyplot as plt

class Perceptron:
    def __init__(self, input_size):
        self.weights = np.random.rand(input_size + 1) - 0.5  # Random initialization between -0.5 and 0.5
        self.learning_rate = 0.1

    def predict(self, inputs):
        summation = np.dot(inputs, self.weights[1:]) + self.weights[0]
        return 1 if summation > 0 else 0

    def train(self, training_inputs, labels):
        while True:
            errors = []
            for inputs, label in zip(training_inputs, labels):
                prediction = self.predict(inputs)
                error = label - prediction
                errors.append(error)
                if error != 0:
                    self.weights[1:] += self.learning_rate * error * inputs
                    self.weights[0] += self.learning_rate * error
            if all(error == 0 for error in errors):
                break

def plot_decision_boundary(perceptron, training_inputs, labels, title):
    plt.figure()
    plt.title(title)
    plt.xlabel('Input 1')
    plt.ylabel('Input 2')
    plt.scatter(training_inputs[:, 0], training_inputs[:, 1], c=labels, cmap=plt.cm.Accent, marker='o')

    w = perceptron.weights
    slope = -w[1] / w[2]
    intercept = -w[0] / w[2]
    x = np.linspace(-0.5, 1.5)
    plt.plot(x, slope * x + intercept, 'k-')

    plt.xlim(-0.5, 1.5)
    plt.ylim(-0.5, 1.5)
    plt.grid(True)
    plt.show()

def boolean_and(inputs):
    return int(all(inputs))

def boolean_or(inputs):
    return int(any(inputs))

def boolean_nand(inputs):
    return int(not all(inputs))

def boolean_xor(inputs):
    return int(sum(inputs) % 2 != 0)

training_inputs = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
labels_and = np.array([boolean_and(inputs) for inputs in training_inputs])
labels_or = np.array([boolean_or(inputs) for inputs in training_inputs])
labels_nand = np.array([boolean_nand(inputs) for inputs in training_inputs])
labels_xor = np.array([boolean_xor(inputs) for inputs in training_inputs])

# Perceptron for AND
and_perceptron = Perceptron(input_size=2)
and_perceptron.train(training_inputs, labels_and)
plot_decision_boundary(and_perceptron, training_inputs, labels_and, 'Decision Boundary for AND')

# Perceptron for OR
or_perceptron = Perceptron(input_size=2)
or_perceptron.train(training_inputs, labels_or)
plot_decision_boundary(or_perceptron, training_inputs, labels_or, 'Decision Boundary for OR')

# Perceptron for NAND
nand_perceptron = Perceptron(input_size=2)
nand_perceptron.train(training_inputs, labels_nand)
plot_decision_boundary(nand_perceptron, training_inputs, labels_nand, 'Decision Boundary for NAND')

# Perceptron for XOR
xor_perceptron = Perceptron(input_size=2)
xor_perceptron.train(training_inputs, labels_xor)
plot_decision_boundary(xor_perceptron, training_inputs, labels_xor, 'Decision Boundary for XOR')
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Function definitions
def f1(x):
    return x**2 - 2*x + 2

def f2(x, y):
    return (1 - x)**2 + 100*(y - x**2)**2

# Gradient descent algorithm
def gradient_descent(f, grad_f, initial_guess, learning_rate=0.01, threshold=1e-5, max_iterations=1000):
    x = initial_guess
    iterations = 0
    while True:
        gradient = grad_f(x)
        x_new = [x_i - learning_rate * grad_i for x_i, grad_i in zip(x, gradient)]
        if np.linalg.norm(np.array(x_new) - np.array(x)) < threshold or iterations >= max_iterations:
            break
        x = x_new
        iterations += 1
    return x, iterations

# Gradient of functions
def grad_f1(x):
    return [2*x[0] - 2]

def grad_f2(x):
    gradient_1 = 2*(200*x[0]**3 - 200*x[0]*x[1] + x[0] - 1)
    gradient_2 = 200*(x[1] - x[0]**2)
    max_gradient = max(abs(gradient_1), abs(gradient_2))
    normalization_factor = max_gradient / 1000  # Choose an appropriate normalization factor
    return [gradient_1 / normalization_factor, gradient_2 / normalization_factor]

# Minimize f(x) = x^2 - 2x + 2
initial_guess_f1 = [0]
minimum_f1, iterations_f1 = gradient_descent(f1, grad_f1, initial_guess_f1)
print("Minimum of f(x) = x^2 - 2x + 2 at x =", minimum_f1, "found in", iterations_f1, "iterations.")

# Minimize f(x, y) = (1 - x)^2 + 100(y - x^2)^2
initial_guess_f2 = [0, 0]
minimum_f2, iterations_f2 = gradient_descent(f2, grad_f2, initial_guess_f2)
print("Minimum of f(x, y) = (1 - x)^2 + 100(y - x^2)^2 at (x, y) =", minimum_f2, "found in", iterations_f2, "iterations.")

# Visualization

# For f(x) = x^2 - 2x + 2
x_vals_f1 = np.linspace(-1, 3, 100)
plt.plot(x_vals_f1, f1(x_vals_f1), label='f(x) = x^2 - 2x + 2')
plt.scatter(minimum_f1[0], f1(minimum_f1[0]), color='red', label='Minimum')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Minimization of f(x) = x^2 - 2x + 2')
plt.legend()
plt.grid(True)
plt.show()

# For f(x, y) = (1 - x)^2 + 100(y - x^2)^2
x_vals_f2 = np.linspace(-2, 2, 100)
y_vals_f2 = np.linspace(-1, 3, 100)
X, Y = np.meshgrid(x_vals_f2, y_vals_f2)
Z = f2(X, Y)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis', alpha=0.8)
ax.scatter(minimum_f2[0], minimum_f2[1], f2(*minimum_f2), color='red', label='Minimum')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('f(x, y)')
ax.set_title('Minimization of f(x, y) = (1 - x)^2 + 100(y - x^2)^2')
plt.legend()
plt.show()

```
